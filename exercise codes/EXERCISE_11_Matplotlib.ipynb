## Part 2

Please recreate the bar chart seen below.  

* The figure should be 8x6
* The two colors used are `#74b9ff` and `#7ed6df`



```python
# Use this data to generate the plot!
resorts = ["Crested Butte", "Aspen", "Telluride", "Wolf Creek"]
snow_last_24 = [29,15,15,23]
snow_next_24 = [16,6,19,30]
```

```python
plt.figure(figsize=(8,6),dpi=100)
plt.bar(resorts, height=snow_last_24, color="#74b9ff",label="Last 24hrs")
plt.bar(resorts, height=snow_next_24, color="#7ed6df",bottom=snow_last_24,label="Next 24hrs")
plt.title("CO Snow Storm Report")
plt.ylabel("Inches")
plt.legend()
plt.show()
```

```python
# Use these lists to generate the chart!
labels = [
    "Building Construction",
    "Deferred Maintenance",
    "General Contractor Fees",
    "Owner Costs",
    "Site Construction"
]
costs = [
    47945975,
    3753584,
    7340612,
    18247137,
    5766685
]
colors = ["#355070", "#6d597a", "#b56576", "#e56b6f", "#eaac8b"]
```

## Part 4

Recreate the figure seen in this image:

* Use the `kc_house_data.csv` dataset (import it with Pandas)
* The figure is 10x4


```python
import pandas as pd
import numpy as np
houses = pd.read_csv("/content/drive/MyDrive/data/kc_house_data.csv",parse_dates=["date"])
week_sale=houses["date"].dt.dayofweek.value_counts()
month_sale=houses["date"].dt.month.value_counts()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.bar(["Mon","Tue","Wed","Thu","Fri","Sat","Sun"],week_sale.sort_index())
plt.title("Sales By Week Day")

plt.subplot(1,2,2)
plt.plot(month_sale.sort_index())
plt.xticks(np.arange(1,13))
plt.title("Sales By Month")
plt.ylabel("Home Sales")
plt.xlabel("Month")
plt.show()
```
